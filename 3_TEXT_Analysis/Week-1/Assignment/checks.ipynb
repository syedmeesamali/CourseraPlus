```python
import pandas as pd

doc = [] #Empty array
labels = ['index', 'text']

with open('dates.txt', 'r') as file:
    for line in file:
        doc.append(line.strip())


file_series = pd.Series(doc)
df = file_series.to_frame()
# print the dataframe
print(df)
```

```python
index = list(range(len(df)))
df['index'] = index
df = df.rename(columns={df.columns[0]: 'text'.strip()})
df.head(5)
```
