## Functions in Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
plt.plot(x, x**3)
plt.plot(x, x**2)
plt.plot(x, x)
plt.show()
```

```python
y = np.linspace(-5, 5, 1000)
plt.plot(y, y**3 - y, c='blue')
plt.plot(y, (y+2)**3 - (y+2), c='red')
plt.plot(y, np.sin(y), c='green')
plt.ylim(-5, 5)
plt.show()
```
