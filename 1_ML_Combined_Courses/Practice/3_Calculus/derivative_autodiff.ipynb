## Derivatives and AutoDiff

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Simple function plot

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
x = np.linspace(-10, 10, 1000)  #Start, finish and # of points
y = f(x)
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x, y)
```

### Let's check slope at x=2

```python
f(2) # y -value when x = 2
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
_ = ax.plot(x, y)
```

In "Delta method" we calculate difference between two points. Let's check another point like x=5.

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(5, f(5), zorder = 3)
_ = ax.plot(x, y)
```

### Slope can be calculated as below.
slope m = (f(5) - f(2) / (5 - 2)) = 9

```python
m = (f(5) - f(2))/ (5 - 2)
b = 37 - m * 5 # y = m*x + b ......therefore b = y - m*x
line_y = m*x + b
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(5, f(5), zorder = 3)
plt.ylim(-10, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x, y)
```

### Here x=2.1 (much closer to x=2)

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(2.1, f(2.1), zorder = 3)
_ = ax.plot(x, y)
```

### Derivative line draw

```python
m = (f(2.1) - f(2))/ (2.1 - 2)
b = f(2.1) - m * 2.1
line_y2 = m*x + b
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(2.1, f(2.1), zorder = 3)
plt.ylim(-5, 150)
plt.plot(x, line_y2, c='orange', zorder = 3)
_ = ax.plot(x, y)
```

# AutoDiff

```python
import torch
x = torch.tensor(5.0)
```

```python
print(x)
```

```python
x.requires_grad_()      #Track gradients through forward pass
```

```python
y = x**2
```

```python
y.backward()
```

```python
x.grad #We will get 10 as  (d/dx (x^2) = 2 * x so at x=5 it will be 10)
```
