## Calculus of Infinitesimals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)  #Start, finish and # of points
```

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-2, -0])
ax.set_ylim([0, 2])
_ = ax.plot(x, y)
```

## Further zoom in on curve to see it flat

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.01, -0.99])
ax.set_ylim([0.99, 1.01])
_ = ax.plot(x, y)
```

## Plot the function again with some graph

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-5, 10)
plt.ylim(-10, 80)
plt.axvline(x=5, color='purple', linestyle='--')
plt.axhline(y=37, color='purple', linestyle='--')
_ = ax.plot(x, y)
```

## Function plot

```python
def my_func(my_x):
    my_y = (my_x**2 - 1) / (my_x - 1)
    return my_y
```

```python
my_func(2)
```

```python
my_func(0.99)
```

```python
my_func(1.01)
```

```python
y = my_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x, y)
```

```python
def sin_func(my_x):
    my_y = np.sin(my_x) / my_x
    return my_y
```

```python
sin_func(0.1)
```

```python
sin_func(-0.1)
```

```python
def my_func2(my_x):
    my_y = 25 / my_x
    return my_y
```

```python
y = my_func2(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
_ = ax.plot(x, y)
```

```python
left_x = x[x < 0]
right_x = x[x > 0]
```

```python
left_y = my_func2(left_x)
right_y = my_func2(right_x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
ax.plot(left_x, left_y, c='C0')
_ = ax.plot(right_x, right_y, c='C0')
```
